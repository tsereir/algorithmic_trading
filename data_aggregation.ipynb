```python
import pandas as pd

def aggregate_timeframe(data, interval):
    """
    Agrège les données 1 minute en un intervalle personnalisé (par exemple, 3 minutes, 5 minutes, etc.).
    """
    aggregated = data.resample(interval).agg({
        'open': 'first',  #1ere val
        'high': 'max',     #val max
        'low': 'min',      # val min
        'close': 'last',   # last val
        'volume': 'sum'    #somme de l'intervalle
    })
    aggregated.dropna(inplace=True)  
    return aggregated
```
