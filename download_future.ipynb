```python
import time as tm
from datetime import time, timedelta, datetime
import schedule
import pandas as pd
import threading
import matplotlib.pyplot as plt
from tqdm import tqdm
from binance.client import Client

client = Client()
```

## Get data from binance

```python
def get_history(client, symbol, interval, duration):
    tm.sleep(0.01)  
    return client.get_historical_klines(symbol=symbol, interval=interval, start_str=duration)
```

```python
test = get_history(client, symbol='BTCUSDT', interval='1m',duration= '60 minutes ago UTC')
```

```python
df = pd.DataFrame(test, columns=['Open Time', 'Open', 'High', 'Low', 'Close', 'Volume', 'Close Time', 'Quote Asset Volume', 'Number of Trades', 'Taker Buy Base Asset Volume', 'Taker Buy Quote Asset Volume', 'Ignore'])
```

```python
df['return'] = (df['Close'].pct_change() + 1).cumprod() - 1            
```

```python
df['date'] = pd.to_datetime(df['Open Time'], unit='ms')

df = df.set_index('date')
df.head(3)
```
